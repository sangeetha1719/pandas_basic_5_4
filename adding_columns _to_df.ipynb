Adding column to a DataFrame (RGLAB)

```python
import numpy as np
import pandas as pd
```

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
        'Score 1' : [56,86,77,45],
        'Score 2' : [50,96,60,30]}

df = pd.DataFrame(data)
```

```python
print('Data Frame Before New Column:')
df.head()
```

```python
address = ['NYC','NJ','CA','PA']

df['Address'] =  address
```

```python
print('Data Frame After New Column:')
df.head()
```

Creating a new column USING an Old column(S)

```python
print('Before column Add:')
df.head()
```

```python
df['Total Score'] = df['Score 1'] + df['Score 2']

print('After Column Add:')
df.head()
```

```python
# Add column using old columns

df['Total scores'] = df['Score 1'] + df['Score 2']

print('After column Add:')
df.head()
```

```python
df['Avg_score'] = df['Total scores'] /2
df.head()
```

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
        'Score 1' : [56,86,77,45],
        'Score 2' : [50,96,60,30]}

df2 = pd.DataFrame(data, columns=['Score 1', 'Score 2', 'Name', 'Qualification'])
print(df2)
```

```python
values = [['Rohan', 455], ['Elvish', 250], ['John', 495],
          ['Sai', 400], ['Eric', 350], ['Adam', 450]]

df2 = pd.DataFrame(values, columns=['Name','Univ_Marks'])
df2.head()
```

```python
## vectorized column
df2['percentage'] = (df2['Univ_Marks'] / 500) * 100
df2.head()
```

Using the .insert()method to add  column

```python
df2.insert(1, 'Age', [1000,2000, 50, 60, 12, 13])
df2.head()
```

Using `.assign()` to control the new column

```python
df2 = df2.assign(location=['MiddleEarth', 'Forgotten Lands', 'NYC', 'NJ', 'CA', 'MA'])
df2.head()
```
