## Handling Missing values
- `df.isnull()` - to detect missing values (isNull = isna() is samething)
- `df.fillna()` - to fill the values ( imputation)
-  `df.dropna()` - to remove missing values
- `df.replace()` - to replace specific values(imputation)

```python
import numpy as np
import pandas as pd
```

```python
employee_list = [['James', 36, 75, 5428000],
               ['Villers', 38, 74, 3428000],
               ['VKole', 31, 70, 8428000],
               ['Smith', 34, 80, 4428000],
               ['Gayle', 40, 100, 4528000],
               ['Adam', 40, np.nan, 4528000],
               ['Rooter', 33, 72, 7028000],
               ['Peterson', 42, 85, 2528000],
               ['lynda', 42, 85, np.nan],
               [np.nan, 42, 85, np.nan],
               ['Jenny', np.nan, 100, 25632],
               ['Kenn', np.nan, 110, 25632],
                ['Aly', np.nan, 90, 25582],
               ['John', 41, 85, 1528000]]

df =pd.DataFrame(employee_list, columns=['Name', 'Age', 'Weight', 'Salary'])
```

```python
# shape
rows,columns = df.shape
print(df.shape)
print(f'Total rows: {rows}, Total columns: {columns}')
print(f'Rows: {rows}')
print(f'Columns: {columns}')
```

```python
# View First 5 Entries
df.head()
```

```python

#Provide a summarization of missing values per column
df.isnull().sum()
```

```python
df.columns
```

```python
# Be wary of missing values , if you have a small dataset, removing values can dramatically alter your analysis
# is missing data Rows is less than 5% of all data it is safe to remove and will have a little effect on analysis
dropped_df = df.dropna()
# when dropping na, you remove the entire row

print(dropped_df)
```

```python
filled_df = df.fillna('None')

filled_df
```

```python
filled_df.isnull().sum()
```

```python
# Slicing and imputing columns at the same time
specific_cols = df[['Weight', 'Age']].fillna('Pending')

specific_cols
```

#### Pandas DATE and TIME
- `pd.to_datetime` - convert different data types into datetime object
- `pd.date_range` - Used to generate a range of data and Time.
- `pd.timestamp` - generate the time expressed as number of seconds since 01/01/1970

```python
# Generate a range of data from start to end , default frequency
date_range = pd.date_range(start='03/01/2026', end='05/05/2026')

print(date_range)
```

```python
# Generate a range of data from start to end , default frequency
# periods is number of dates yu want between start and end (inclusive)
date_range = pd.date_range(start='03/01/2026', end='05/05/2026',periods=4)

print(date_range)
```

```python
start_date = '01/01/2023'
num_days=30

date_range = pd.date_range(start=start_date, periods=num_days, freq='D')

np.random.seed(42) # seed means we are all using the SAME random algorithm
# mean - 100
# std - 5
# 30 numbers
stock_prices = np.random.normal(loc=100, scale=5, size=num_days).round(2)

stock_df = pd.DataFrame({'Date': date_range, 'StockPrice':stock_prices})

stock_df
```

```python
url='https://raw.githubusercontent.com/bprasad26/lwd/master/data/tesla_stock_prices.csv'
# Using parse dates to change type from string info date
tesla_df = pd.read_csv(url, parse_dates=['Date'])

# tesla_df.to_csv('data/tesla_stocks.csv')
tesla_df.head()
#tesla_df.info
```

```python
# Add a new columns to dataframe based off of dates
tesla_df['Month'] = tesla_df['Date'].dt.month
tesla_df['Year'] = tesla_df['Date'].dt.year
tesla_df['Day'] = tesla_df['Date'].dt.day
tesla_df['day_name'] = tesla_df['Date'].dt.day_name()

tesla_df.head(10)
```

```python
# grab max or min date from a dataset
most_recent_date = tesla_df['Date'].max()   # newest date
least_recent_date = tesla_df['Date'].min()   # oldest date

print(f'Most Recent Date: {most_recent_date}')
print(f'Least Recent Date: {least_recent_date}')
```

```python
# Group by month and year and sum of the volume
# Groups your data by Year and Month combinations-
monthly_sales = tesla_df.groupby(['Year', 'Month'])['Volume'].sum().reset_index()

monthly_sales.head(15)
```
