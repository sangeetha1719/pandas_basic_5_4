### Data Frame - Index and column

```python
fruit_data = {
    'apple_sales': [3,2,3,5,7,0,2,2],
    'orange_sales': [3,2,5,5,8,0,2,2]
}
# idx = ['Day 1','Day 2','Day 3','Day 4','Day 5','Day 6','Day 7','Day 8']

idx = pd.RangeIndex(2000, 2008)
purchases = pd.DataFrame(fruit_data, index=[idx])
print(purchases)
```

```python
rows, cols = purchases.shape
print(f'Rows: {rows}')
print(f'Columns: {cols}')
```

```python
purchases.head() #view first 5 elements
```

```python
purchases.tail() # view bottom 5 elements
```
