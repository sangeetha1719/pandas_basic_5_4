Sort Using Pandas --- 5/5/2026

```python
data = {'Name': ['Jane', 'Princi', 'James', 'Fadi'],
        'Height': [5.1, 6.2, 5.1, 5.2],
        'Qualification': ['Msc', 'MA', 'Msc', 'Msc'],
       'Score 1' : [56,86,77,45],
       'Score 2' : [50,96,60,30]}
grades_df = pd.DataFrame(data)

grades_df.head()
```

```python
grades_df = ([grades_df.assign(address ='NYC', 'NJ', 'CA','PA')])

grades_df.head()
```

```python
#sort _values().sortby a ascending order
print(grades_df.sort_values(by='Score 1'))
```

```python
print(grades_df.sort_values(by='Score 1', ascending=False))
```

```python
cars = pd.read_json('data/cars.json')

cars.head(15)
```

```python
print(cars.sort_values(by ='quantity', ascending = True))
```

```python
# sort values by qty and by car
print(cars.sort_values ( by = ['quantity','car'] ascending = [False,True]))
```

Locate and ilocate
### Slicing a dataframe with .iloc[] & .loc[]
-  'df.loc[row:row , col:col]' - local based indexing
- `df.iloc[row:row, colNumber:colNumber] - indexbasedlocation

```python
# Initializing the nested list with Data set
employee_list = [['James', 36, 75, 5428000],
               ['Villers', 38, 74, 3428000],
               ['VKole', 31, 70, 8428000],
               ['Smith', 34, 80, 4428000],
               ['Gayle', 40, 100, 4528000],
               ['Rooter', 33, 72, 7028000],
               ['Peterson', 42, 85, 2528000],
               ['John', 41, 85, 1528000],

]
e_df = pd.DataFrame(employee_list, columns=['Name', 'Age', 'Weight', 'Salary'])

e_df.head()
```

```python
# first row
e_df.iloc[0]
```

```python
# Grabs the first row
e_df.iloc[-1]
```

```python
# grabs the first row
e_df.iloc[:1]
```

```python
e_df.iloc[0:4]
```

```python
# grabs all rows but only the first three columns
e_df.iloc[: , 0:3]
```

```python
# grab first two rows and first two columns
e_df.iloc[0:2,0:2]
```

```python
basketball = {'team': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
                   'points': [18, 22, 19, 14, 14, 11, 20, 28],
                   'assists': [5, 7, 7, 9, 12, 9, 9, 4],
                   'rebounds': [11, 8, 10, 6, 6, 5, 9, 12],
                   'steals': [4, 3, 3, 2, 5, 4, 3, 8],
                   'blocks': [1, 0, 0, 3, 2, 2, 1, 5]}

ball_up = pd.DataFrame(basketball)
ball_up.head()
```

```python
# locate all rows and columns team, rebounds
ball_up.loc[:, ['team', 'rebounds']]
```

```python
# select values from row 3 -6 and show the team name
ball_up.loc[3:6, ['team']]
```
