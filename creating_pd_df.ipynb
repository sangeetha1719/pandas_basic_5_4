### Creating a Pandas DataFrame

```python
import numpy as np
import pandas as pd
```

```python
# create pandas df from a dictionary
d = {'x': [1, 2, 3], 'y': [2, 4, 8], 'z': 100}
df = pd.DataFrame(d)
print(f'dictionary: d')
print("\nDataFrame")
print(df)
```

```python
# make dataframe have an index [100, 200, 300]
# make the columns names [z, x, y]
df2 = pd.DataFrame(d, index=[100, 200, 300], columns=['z', 'y', 'x'])

df2
```

```python
student_dict = {
    'Name': ['Joe', 'Nat', 'Harry'],
    'Age': [20, 21, 19],
    'Marks': [85.10, 77.80, 91.54]
}
student_df = pd.DataFrame(student_dict)
```

```python
student_df = student_df.set_index('Name')
print(student_df.head())
```

```python
data = {
    'Product': ['Laptop', 'Tablet', 'Phone'],
    'Price': [1200, 300, 800],
    'Quantity': [50, 150, 100]
}
df3=pd.DataFrame(data)

df3.head()
```

```python
df3.index = pd.RangeIndex(1,4,1)
df3.head()
```

### Creating a df using Zip() function

```python
# create a list of patientID, name, and date of birth and assign it to a variable
patientID = [101,23,48,49]
name =       ['alice','bob','charlie','Eric']
# create a list of dates
date_of_birth = ['2023-01-01', '2023-01-02', '3/10/2020 143045', '13th of October, 2023']
# when using zip() need to create columns' name
employee_df = pd.DataFrame(zip(patientID,name,date_of_birth), columns=['patientID','name','dob'])

employee_df
```

```python
# in this example, we create a list of stocks and their prices and then create a DataFrame.
stocks = ["IBM", "APPLE", "TWTTR", "GE", "MSFT"]
prices = [115.00, 119.14, 19.77, 25.99, 26]

# stock_df1 = pd.DataFrame(list(zip(stocks,prices)), columns=['stock','price'])

stock_df = pd.DataFrame(zip(stocks,prices), columns=['stock','price'])

stock_df
```

```python
arr = np.array([[1,2,100],[2,4,100],[3,8,100]])
numpy_df = pd.DataFrame(arr, columns=['x','y','z'])
numpy_df.head()
```
